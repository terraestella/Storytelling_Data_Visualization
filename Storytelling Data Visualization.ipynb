## Storytelling Data Visualization on Exchange Rates ##

**Showcase**: explanatory data visualization with matplotlib

__Dataset__: exchange rates between 1999 and 2022 [link](https://www.kaggle.com/lsind18/euro-exchange-daily-rates-19992020)

### Dataset Import and Initial Exploration ###

```python
import pandas as pd
exchange_rates = pd.read_csv('euro-daily-hist_1999_2022.csv')
```

```python
exchange_rates.shape
```

```python
pd.set_option('display.max_columns',41)
exchange_rates.head()
```

```python
exchange_rates.info()
```

### Data Cleaning ###

```python
# renaming columns (simplifying columns names)
exchange_rates.rename(columns={'[US dollar ]': 'US_dollar',
                               'Period\\Unit:': 'Time'}, inplace=True)

exchange_rates
```

```python
# converting Time column to datetime data type
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])

# sorting the dataset values by Time column in ascending order, resetting index and dropping initial index
exchange_rates.sort_values('Time', inplace=True)
exchange_rates.reset_index(drop=True, inplace=True)

exchange_rates
```

```python
# extracting Time and US_dollar columns
euro_to_dollar = exchange_rates[['Time', 'US_dollar']].copy()

euro_to_dollar
```

```python
# exploring US_dollar column value counts
euro_to_dollar['US_dollar'].value_counts()
```

There are 62 dash characters in that column.

```python
# dropping dashes from US_dollar column and converting it to float data type
euro_to_dollar = euro_to_dollar.copy()[euro_to_dollar['US_dollar'] != '-']
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)

euro_to_dollar.info()
```

### Visualization ###

```python
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline

plt.plot(euro_to_dollar['Time'],
         euro_to_dollar['US_dollar'])
plt.show()
```

To get rid off all the daily changes in the graph, making the line smoother and easier to interpret, we will show only the long-term trends - **the rolling mean** (a.k.a. the moving average). 

```python
# the rolling means for the US_dollar column, using a 30-day moving average
euro_to_dollar['rolling_mean'] = euro_to_dollar['US_dollar'].rolling(30).mean()

euro_to_dollar
```

```python
plt.plot(euro_to_dollar['Time'],
         euro_to_dollar['rolling_mean'])
plt.show()
```

### Storytelling Data Visualization Example #1: The Effects of Covid-19 Pandemic on Eur/USD Exchange Rates ###

```python
eur_usd_covid = euro_to_dollar[euro_to_dollar['Time'].dt.year >= 2019]
eur_usd_covid
```

```python
import matplotlib.style as style
%matplotlib inline

# showing available plotting styles in matplotlib.style
style.available
```

```python
# plt.rcParams.update(plt.rcParamsDefault) - reverts the stylesheet (deploy when switching between styles)
```

```python
# choosing the style
style.use('fivethirtyeight')

# plotting the Time/US_dollar values
fig, ax = plt.subplots(figsize=(8,3))
ax.plot(euro_to_dollar['Time'],
        euro_to_dollar['rolling_mean'],
        linewidth=2)

# highlighting the pandemic period 
ax.plot(eur_usd_covid['Time'],
        eur_usd_covid['rolling_mean'],
        linewidth=3, color='#b22222')
ax.set_title('Eur/USD Exchange Rates During Covid-19 Pandemic', fontsize=12)
plt.show()
```

```python
# first case in US: 2020-01-19
covid_us = eur_usd_covid[eur_usd_covid['Time'] >= '2020-01-19']

fig, ax = plt.subplots(figsize=(12,3))
ax.plot(eur_usd_covid['Time'], eur_usd_covid['US_dollar'].rolling(30).mean())
plt.xticks(rotation=30)

ax.plot(covid_us['Time'], covid_us['US_dollar'].rolling(30).mean(),
        linewidth=5, color='#b22222')

ax.set_title('Eur/USD Exchange Rate Since First Case in US', fontsize=12)

plt.show()
```

### Storytelling Data Visualization Example #2: Eur/USD Exchange Rate During the 3 Different US Presidents ###

```python
# Bush-Obama-Trump years
bot = euro_to_dollar[(euro_to_dollar['Time'].dt.year >= 2001) & (euro_to_dollar['Time'].dt.year < 2021)]
# the Bush administration (2001-2009)
bush = bot[bot['Time'].dt.year < 2009]
# the Obama administration (2009-2017)
obama = bot[(bot['Time'].dt.year >= 2009) & (bot['Time'].dt.year < 2017)]
# the Trump administration (2017-2021)
trump = bot[(bot['Time'].dt.year >= 2017) & (bot['Time'].dt.year < 2021)]
```

```python
# creating subplots
plt.figure(figsize=(12,6))
ax1 = plt.subplot(2,3,1)
ax2 = plt.subplot(2,3,2)
ax3 = plt.subplot(2,3,3)
ax4 = plt.subplot(2,1,2)
axes = [ax1, ax2, ax3, ax4]

# all subplots
for ax in axes:
    ax.set_ylim(0.8, 1.7)
    ax.set_yticks([1.0, 1.2, 1.4, 1.6])
    ax.set_yticklabels(['1.0', '1.2', '1.4', '1.6'], alpha=0.3)
    ax.grid(alpha=0.5)
    
# bush subplot
ax1.plot(bush['Time'], bush['rolling_mean'], color='red')
ax1.set_xticklabels(['','2001', '', '2003', '', '2005', '', '2007', '', '2009'], alpha=0.3)
ax1.set_title("BUSH\n(2001-2009)", fontsize=11, color='red', weight='bold')

# obama subplot
ax2.plot(obama['Time'], obama['rolling_mean'], color='blue')
ax2.set_xticklabels(['', '2009', '', '2011', '', '2013', '', '2015', '', '2017'], alpha=0.3)
ax2.set_title("OBAMA\n(2009-2017)", fontsize=11, color='blue', weight='bold')

# trump subplot
ax3.plot(trump['Time'], trump['rolling_mean'], color='red')
ax3.set_xticklabels(['', '2017', '', '2018', '', '2019', '', '2020', '', '2021'], alpha=0.3)
ax3.set_title("TRUMP\n(2017-2021)", fontsize=11, color='red', weight='bold')

# all together
ax4.plot(bush['Time'], bush['rolling_mean'], color='red')
ax4.plot(obama['Time'], obama['rolling_mean'], color='blue')
ax4.plot(trump['Time'], trump['rolling_mean'], color='red')
ax4.set_xticklabels([])

# title
plt.title("Eur/USD Exchange Rate During the 3 Different US President",
          x=0.5, y=2.5, fontsize=18, weight='bold')
plt.suptitle("(Source: European Central Bank)", fontsize=10)

plt.show()
```
